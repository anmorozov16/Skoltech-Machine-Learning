$$\sum_{i=1}^{n}v_{i}\cdot L(y_{i}, \hat{y}_{i})+\frac{\lambda}{2}\|w\|_{2}^{2}=\sum_{i=1}^{n}v_{i}\cdot (\boldsymbol{w}\cdot\boldsymbol{x}_{i}-y_{i})^{2}+\frac{\lambda}{2}\|w\|_{2}^{2}\rightarrow \min_{\boldsymbol{w}}.$$

where $V=V^{\top}\in\mathbb{R}^{n\times n}$ is the diagonal matrix with diagonal elements $v_{1},\dots, v_{n}$. ## Task 2. Logistic Regression (1 point)
Let us consider the case when in the problem of binary classification the training set is lineary separable. Show that in this case the optimization problem for logistic regression **without L2-regularization** does not have optimum.
### Your solution:

Logarithmic loss function with binary logistic regression with no regulatization:

$$\overline{L}(w)=\sum_{i=1}^{m}\log(1+\exp(-(w^{\top}x_{i})y_{i}))\rightarrow \min_{w}$$

Linear separation condition:

$H = {x → sgn(w · x{\top} + b), w \in R^{d}, b \in R}$

Interpretation:

$y_{i}(w^{\top}x_{i})>0$ i.e.:


$y \in [-1, +1]$

$y_{i}=1$, $w^{\top}x_{i}+b>>0$

$y_{i}=-1$, $w^{\top}x_{i}+b<<0$

Therefore, $y_{i}(w^{\top}x_{i}+b)>0$ when classified correctly when signs from $y_{i}$ and $w^{\top}x_{i}$ are the same


Taking a partial derivative with respect to w:

$$\frac{\partial \overline{L}}{\partial w}=\sum_{i=1}^{m}\frac{-\exp\big[{-(w^{\top}x_{i})y_{i}}\big]y_{i}x_{i}}{{1+\exp{\big[-(w^{\top}x_{i})y_{i}\big]}}}\neq 0 $$

The $\exp\big[{-(w^{\top}x_{i})y_{i}}\big]$ on the numerator and denominator cannot be zero, $x_{i}, y_{i}$ are positive from the classification condition.

Hence, the logistic regression with no L2-regularization has no optimum.

## Task 3. Bayesian Naive Classifier (1 point)
Show that in case of $d$ binary-valued features $x_{j}\in\{0, 1\}$ (for $j=1,2,\dots,d$) Bayesian Naive Classifier's decision rule is linear.
### Your solution:

Naive Bayes Classifier:
$$p(C_{k} | x) \sim p(x | C_{k}) \cdot p(C_{k})\tag{1}$$

We utilize Bernoulli NB since its features are independent binary inputs:

$$p(x|C_k)=\prod_{i=1}^{d}p_{ki}^{x_i}(1-p_{ki})^{(1-x_i)}\tag{2}$$

Combining (1) and (2) and taking log:

\begin{align}
    \log{p(x|C_k)}
        &= \log\prod_{i=1}^{d}\Big(p_{ki}^{x_i}(1-p_{ki})^{(1-x_i)}\Big)+\log{p(C_{k})}
        \\
        &= \sum_{i=1}^{d}\Big(x_i\log p_{ki}+(1-x_i)\log(1-p_{ki})\Big)+\log{p(C_{k})}
        \\
        &= \sum_{i=1}^{d}\Big(x_i\log p_{ki}+\log(1-p_{ki})-x_{i}\log(1-p_{ki})\Big)+\log{p(C_{k})}
        \\
        &= \sum_{i=1}^{d}\Big(x_i\log p_{ki}-x_i\log(1-p_{ki})\Big)+\sum_{i=1}^{d}\log(1-p_{ki})+\log{p(C_{k})}
\end{align}

The first term $\sum_{i=1}^{d}\Big(x_i\log p_{ki}-x_i\log(1-p_{ki})\Big)$ can be interpreted as $x^{\top}\alpha$

The rest terms $\sum_{i=1}^{d}\log(1-p_{ki})+\log{p(C_{k})}$ are just some constant, let's say $\beta$

Hence, the NB decision rule equation is also linear $x^{\top}\alpha+\beta$:

$$
\hat{y} = \arg \max_{k} P(C_k | x) = \arg \max_{k} (x^{\top}\alpha+\beta)
$$

## Task 8. Kernel Regression (1 point)
Recall that the prediction of Kernel Ridge Regression fitted on data $X^{n}$ with the kernel $K(\cdot, \cdot)$ has the form $\mathcal{K}(x)=\sum_{i=1}^{n}\alpha_{i}K(x, x_{i})$, where $\alpha=(K+\lambda I)^{-1}Y$ ($K_{ij}=K(x_{i},x_{j})$). The time complexity of computation of a prediction $\mathcal{K}(x)$ for any point $x$ is $O(n)$, i.e. grows linearly with the size of the training set.

Consider the bilinear kernel $K(x, x')=\langle x, x'\rangle$. For this kernel, the Kernel Regression is known to turn into simple linear ridge regression. However, for linear regression the computation time of prediction $\mathcal{R}(x)=\sum_{j=1}^{d}\beta_{j}x^{j}$ is $O(d)$, where $d$ is the dimension of the feature space and does not grow with the training, which is a little bit controversary to the previous paragraph.

In this task in order to show that the kernel regression with the bilinear kernel is indeed the linear ridge regression, you have to prove that the predictions exactly match by **direct comparison**.
### Your solution:

Recall the objective function for the ridge regression:

$$ F(W) = ||XW-Y||^2 + \lambda||W||^2 \tag{1}$$

The solution for parameter $W$ is via $\nabla_{w} F(w)= 0$:

$$W=(X^{\top}X + {\lambda}I)^{-1}X^{\top} Y\tag{2}$$ 

By introducing new variable $\alpha=(XX^{\top} + {\lambda}I)^{-1}Y$, we obtain:

$$W=X^{\top}\alpha=\sum_{i=1}^{m}\alpha_{i} x_{i}^{\top}=\sum_{i=1}^{m}\alpha_{i}\langle x', x_{i} \rangle$$

Taking into account (3), the solution is:

$$\hat{f}(x')=x'\cdot W=\sum_{i=1}^{m}\alpha_{i}\langle x', x_{i} \rangle=x'x^{\top}(xx^{\top} + {\lambda}I)^{-1}Y \tag{4}$$

At the same time, the Kernel $K\langle x, x'\rangle$ with the given $\alpha=(K+\lambda I)^{-1}Y$($K_{ij}=K(x_{i},x_{j})$) ridge regression is the following:

\begin{align}
    \hat{f}(x')
        &=\sum_{i=1}^{m}\alpha_{i}K\langle x', x_{i}\rangle
        \\
        &= \sum_{i=1}^{m}\alpha_{i}\langle x', x_{i}\rangle
        \\
        &= \sum_{i=1}^{m}x'x_{i}^{\top}\alpha_{i}
        \\
        &= x'x^{\top}(K+\lambda I)^{-1}Y
        \\
        &=x'x^{\top}(xx^{\top} + {\lambda}I)^{-1}Y \tag{5}
\end{align}

Hence, (4) from the linear ridge regression is the same as (5) from the kernel regression with the bilinear kernel.

By using the above mentioned kernel properties, we calculate the following:

$$K(x,x')=\exp(-\|x-x'\|^{2})=\exp(-\|x\|^{2})\exp(2x^{\top}|x'|)\exp(-\|x'\|^{2})$$

A product of kernels is a valid kernel according to the mentioned property 1. Obviously, the terms $\exp(-\|x\|^{2})$ and $\exp(-\|x'\|^{2})$ cannot be negative due to exponential function, and the terms are valid according to the property 2.
By refering to the third kernel property, the product of kernels is a valid kernel, and for the second term $\exp(2x^{\top}x)$ this property gives $\exp{k(x,x')}$. In its turn, this kernel is also positive definite since $x^{\top}|x'|$ is positive difinite.
