## Task 1. Numpy Problem 1 (1 point)

Write a function, which takes a matrix `arr` and centers each of its rows
by the mean within that row. Check [this](https://docs.scipy.org/doc/numpy/reference/routines.statistics.html)
out for documentation.

```python
def center(arr):
    """Center each row of a matrix by the mean value in that row.

    Parameters
    ----------
    arr : array, shape = (n_rows, n_cols)
        The matrix, the rows of which to center.

    Returns
    ----------
    out : array, shape = (n_rows, n_cols)
        The final row-centered matrix.
    """
    assert arr.ndim == 2
    n_rows, n_cols = arr.shape
    mean = arr.mean(axis=1)
    out = arr.copy().astype(float)
    for m, i in zip(mean, out):
        m -= i
    return out
```

## Task 2. Numpy Problem 2 (1 point)

Implement the following fancy function:
\begin{equation}
    f(x) = \sigma\bigl(
                \max\{x + 5, 0\} + \max\{5 - x, 0\}
                + \max\{\min\{\cos(2 x \pi), \tfrac12\}, -\tfrac14\}
            \bigr)
         \,,
\end{equation}
where $\sigma(x) = (1+e^{-x})^{-1}$ is the sigmoid function.

```python
def fancy_function(x):
    """Compute some fancy function.
    
    Parameters
    ----------
    x : array, 1 dimendional, shape=(n_samples,)
        The array argument values.

    Returns
    -------
    y : array, 1 dimendional, shape=(n_samples,)
        The values of the fancy function.
    """
    max1 = np.maximum(x+5, 0)
    max2 = np.maximum(5-x, 0)
    max3 = np.maximum(np.minimum(np.cos(2*np.pi*x), .5), -.25)
    c = max1 + max2 + max3
    out = 1 / (1 + np.exp(-c))
    
    return out
```

Plot the function

### Subproblem 5.2. Feature engeneering for Random Forest  (1 of 2 points)
Perform **feature engeneering** for Random Forest in order to achieve 5-fold validation score not lower than $0.94$. Write 2-3 sentences to explicitly explain your motivation for provided feature choice.

```python
def cr():
    val = X.values
    val = val.reshape((val.shape[0], 3, 3))
    g = np.empty((8,))
    xx = np.empty((val.shape[0], 1))
    for j, v in enumerate(val):
        g[:3] = np.sum(v, axis=0)
        g[3:6] = np.sum(v, axis=1)
        g[6] = v[0, 0] + v[1, 1] + v[2, 2]
        g[7] = v[0, 2] + v[1, 1] + v[2, 0]
        xx[j, 0] = np.sum(g==3, axis=0)
    return xx
x = cr()
rf = RandomForestClassifier(n_estimators=400, max_depth=10, random_state=404)
params = {'n_estimators': [200, 300, 400],
          'max_depth': [2, 5, 10]}

rfgrid = GridSearchCV(rf, params, cv=5)
rfgrid.fit(x, y)
print(rfgrid.best_params_)
print(rfgrid.best_score_)
```

A feature is created which calculates a sum of `x` per row, per column or on the diagonal. Basically, this feature shows if a player wins or not according to the Tic-Tac-Toe game. If either sum from the row, column or diagonal is equal to 3 `x`'s, then player wins. That is why the accuracy is 1.

Please note that you also have to choose the metric to optimize during CV (e.g. accuracy, balanced accuracy) on your own.

```python
clfs = [LogisticRegression(),
        RandomForestClassifier(n_estimators=100, max_depth=3),
        KNeighborsClassifier(n_neighbors=5),
        DecisionTreeClassifier(max_depth=7), 
       ]

def dot(fun):
    def d(clf, x, y):
        lg = fun(clf)
        lg.fit(x, y)
        return balanced_accuracy_score(y, lg.predict(x))
    return d

@dot
def Multiclass(clf):
    return clf

@dot
def OneVsRest(clf):
    return OneVsRestClassifier(clf)

@dot
def OneVsOne(clf):
    return OneVsOneClassifier(clf)

def func(i, j, _):
    ls = [Multiclass, OneVsRest, OneVsOne]
    for row in range(j.shape[0]):
        for column in range(j.shape[1]):
            time_begin = time()
            d = ls[row](clfs[column], X, y)
            time_end = time()
            i[row][column] = np.array([d, time_end - time_begin])
    return i

df = np.fromfunction(func, (3, len(clfs), 2))

columns = ['Logistic', 'Random Forest', 'K Nearest Neighbors', 'Decision Tree']
rows = ['Multiclass', 'OneVsRest', 'OneVsOne']
CVS_df = pd.DataFrame(df[:, :, 0], columns=columns, index=rows)
TD_df = pd.DataFrame(df[:, :, 1], columns=columns, index=rows)

print('Cross Validation Scores with Balanced Accuracy Score')
display(CVS_df)
print('Training Duration (CV=5)')
display(TD_df)
```
